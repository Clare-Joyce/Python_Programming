```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
flights = pd.read_csv("flights_data.csv")
print(flights.shape)
flights.head()
```

# **Bar Charts**

```python
sns.countplot(x='Source', data=flights)
plt.xlabel('Source', fontsize=12)
plt.ylabel("Number of flights", fontsize=12);
```

```python
sns.countplot(x='Source', data=flights, color=sns.color_palette()[0])
plt.xticks(rotation=30);
```

```python
sns.countplot(x='Source', data=flights, 
              color=sns.color_palette()[1],
              order= flights.Source.value_counts().index);
```

```python
sns.countplot(x='Airline', data=flights,
              color=sns.color_palette()[2]);
```

```python
sns.countplot(x='Airline', data=flights,
              color=sns.color_palette()[2])
plt.xticks(rotation=90);
```

```python
sns.countplot(y='Airline', data=flights,
              color=sns.color_palette()[2]);
```

## **Count Missing Data**

```python
flights.isnull().sum()
```

```python
missing_value_counts = flights.isnull().sum()
sns.barplot(x=missing_value_counts.index.values,
            y=missing_value_counts,
            color=sns.color_palette()[0])
plt.ylabel("Number of missing values", fontsize=12)
plt.xticks(rotation=90);
```

```python
destination_counts = flights.Destination.value_counts()
plt.pie(destination_counts, 
        labels=destination_counts.index,
        startangle=90,
        counterclock=False)
plt.axis("square")
plt.title("Flight Destination's");
```

```python
plt.pie(destination_counts, 
        labels=destination_counts.index,
        startangle=90,
        counterclock=False,
        wedgeprops={'width':0.4})
plt.axis("square");
# plt.title("Flight Destination's");
```

## **Histograms**
