### Exercise 2###
Clean data and update the CSV file. Replace all column values which contain ‘?’ and n.a with NaN.

````python
data_auto = data_auto.replace(r'^\s*?$', np.nan, regex=True)
data_auto = data_auto.replace('n.a', np.nan, regex=True)
````

````python
data_auto.head()
````

````python
data_auto.to_csv('Automobile_data.csv')
````

### Exercise 9 ###
Concatenate two data frames using following conditions
Create two data frames using following two Dicts, Concatenate those two data frame and create a key for each data frame.
```
* GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
* JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

````python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
````

````python
JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
````

````python
GermanCars = pd.DataFrame(GermanCars)
JapaneseCars = pd.DataFrame(JapaneseCars)
````

````python
JapaneseCars
````

````python
result = pd.concat([GermanCars, JapaneseCars], keys = ["Germany", "Japan"])
````

````python
result
````

### Exercise 10 ###
Merge two data frames using following condition.
Create two data frames using following two Dicts, Merge two data frames, and append second data frame as a new column to first data frame.

```
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

````python
Car_Price = pd.DataFrame({'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]})
car_Horsepower = pd.DataFrame({'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]})
````

````python
merged = pd.merge(Car_Price, car_Horsepower, on = 'Company')
merged
````
